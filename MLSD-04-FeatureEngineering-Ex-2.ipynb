## Target/Mean Encoding

```python
# Import libraries
import pandas as pd
```

```python
# Create new dataset
data = {'Customer':['C1','C2','C3','C1','C4','C3','C2','C1','C2','C4','C1','C5','C6','C4','C5','C6'],
        'Target'  :[1,0,1,1,1,0,0,1,1,1,0,1,1,1,1,0]}
df_tme  = pd.DataFrame(data)
print(df_tme)
```

```python
df_tme
```

```python
# Determine the count values of 'Target' column based on the category in 'Customer' variable
# Write code here

df_tme.groupby('Customer').count()
```

```python
# Determine the mean values of 'Target' column based on the category in 'Customer' variable
# Write code here

means = df_tme.groupby('Customer')['Target'].mean()
```

```python
# Display the encoded predictor
# Write code here

df_tme['Customer_encoded'] = df_tme['Customer'].map(means)
df_tme
```
