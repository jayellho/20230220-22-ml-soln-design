## Missing Value Indicator

```python
# Create new column 'Age_NA', column has 1 if the 'Age' column observation is missing, or 0 otherwise.
# Write code here

df['Age_NA'] = df['Age'].isnull().astype(int)
```

```python
# Display first ten values of 'Age' and 'Age_NA' columns
# Write code here

df[['Age','Age_NA']].head(10)
```

**Observations**<br>
- Age in row 6 is NaN.
- Age_NA variable is created to capture the missingness.

## Median Data Imputation

```python
# Fill in median value of Age in Age variable where value is NaN
# Display the first ten rows of df
# Write code here

df['Age'] = df['Age'].fillna(df['Age'].median())

df.head(10)
```

## One-hot Encoding

```python
# Display first 5 rows of Sex variable
# Write code here
new_df = pd.DataFrame(df['Sex'])
new_df.head(5)
```

```python
# One-hot encode Sex variable and display first 5 rows of the two encoded columns
# Write code here
onehotenc_sex_df = pd.get_dummies(df['Sex'])

onehotenc_sex_df.head(5)

```

```python
# Concatenate Sex variable with the newly created encoded columns and display as shown
# Write code here

new_concat_df = pd.concat([new_df, onehotenc_sex_df], axis=1)

new_concat_df
```
